```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression
from sklearn.preprocessing import LabelEncoder


def feature_selection(df, target_column=None, threshold_corr=0.9, missing_threshold=0.4):
    """
    Универсальная функция для выбора важных признаков в датафрейме.
    
    Параметры:
    df - pandas DataFrame
    target_column - колонка с целевой переменной (если есть)
    threshold_corr - порог корреляции (удаляются сильно коррелирующие признаки)
    missing_threshold - порог удаления признаков с большим количеством пропусков
    
    Возвращает:
    Список оставшихся признаков
    """
    df = df.copy()
    
    # 1. Удаление признаков с большим количеством пропущенных значений
    missing_ratio = df.isnull().sum() / len(df)
    cols_to_drop = missing_ratio[missing_ratio > missing_threshold].index.tolist()
    df.drop(columns=cols_to_drop, inplace=True)
    
    # 2. Удаление признаков с низкой дисперсией
    low_variance_cols = df.var()[df.var() == 0].index.tolist()
    df.drop(columns=low_variance_cols, inplace=True)
    
    # 3. Удаление коррелированных признаков
    corr_matrix = df.corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    high_corr_features = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold_corr)]
    df.drop(columns=high_corr_features, inplace=True)
    
    # 4. Выбор наиболее значимых признаков (если есть целевая переменная)
    if target_column and target_column in df.columns:
        X = df.drop(columns=[target_column])
        y = df[target_column]
        
        # Кодирование категориальных признаков
        X = X.apply(lambda col: LabelEncoder().fit_transform(col) if col.dtype == 'object' else col)
        
        if y.dtype == 'object':  # Классификация
            y = LabelEncoder().fit_transform(y)
            importance = mutual_info_classif(X, y)
        else:  # Регрессия
            importance = mutual_info_regression(X, y)
        
        feature_importance = pd.Series(importance, index=X.columns)
        important_features = feature_importance[feature_importance > 0].index.tolist()
        df = df[important_features + [target_column]]
    
    return df
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression

def preprocess_dataframe(df, target_column=None, missing_threshold=0.5, corr_threshold=0.9, pca_components=None, z_threshold=3):
    df = df.copy()
    
    # 1. Удаление столбцов с большим количеством пропущенных значений
    missing_ratio = df.isnull().mean()
    df = df.loc[:, missing_ratio < missing_threshold]
    
    # 2. Удаление признаков с нулевой дисперсией
    df = df.loc[:, df.nunique() > 1]

    # 3. Удаление высоко коррелированных признаков
    numeric_df = df.select_dtypes(include=[np.number])
    corr_matrix = numeric_df.corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > corr_threshold)]
    df = df.drop(columns=to_drop, errors="ignore")
    
    # 4. Определение важных признаков (если указан target_column)
    if target_column and target_column in df.columns:
        X = df.drop(columns=[target_column])
        y = df[target_column]
        if y.nunique() > 2:  # Регрессия
            importance = mutual_info_regression(X.fillna(0), y)
        else:  # Классификация
            importance = mutual_info_classif(X.fillna(0), y)
        important_features = X.columns[np.argsort(importance)[-10:]]  # Топ-10
        df = df[[*important_features, target_column]]
    
    # 5. Нормализация данных
    scaler = StandardScaler()
    df[df.select_dtypes(include=[np.number]).columns] = scaler.fit_transform(df.select_dtypes(include=[np.number]))
    
    # 6. PCA (если указано количество компонент)
    if pca_components:
        pca = PCA(n_components=pca_components)
        pca_data = pca.fit_transform(df.select_dtypes(include=[np.number]))
        df = pd.DataFrame(pca_data, columns=[f"PCA_{i+1}" for i in range(pca_components)])
    
    # 7. Удаление выбросов с помощью Z-score
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    z_scores = np.abs((df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std())
    df = df[(z_scores < z_threshold).all(axis=1)]
    
    return df


```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression

def preprocess_dataframe_exp(df, target_column=None, missing_threshold=0.5, corr_threshold=0.9, pca_components=None, z_threshold=3):
    df = df.copy()
    
    # 1. Remove columns with excessive missing values
    missing_ratio = df.isnull().mean()
    df = df.loc[:, missing_ratio < missing_threshold]
    
    # 2. Remove zero-variance features
    df = df.loc[:, df.nunique() > 1]

    # 3. Remove highly correlated features
    numeric_df = df.select_dtypes(include=[np.number])
    corr_matrix = numeric_df.corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > corr_threshold)]
    df = df.drop(columns=to_drop, errors="ignore")
    
    # 4. Select important features based on mutual information
    if target_column and target_column in df.columns:
        X = df.drop(columns=[target_column])
        y = df[target_column]
        if y.nunique() > 2:  # Regression
            importance = mutual_info_regression(X.fillna(0), y)
        else:  # Classification
            importance = mutual_info_classif(X.fillna(0), y)
        important_features = X.columns[np.argsort(importance)[-10:]]  # Top-10 features
        df = df[[*important_features, target_column]]
    
    # 5. Normalize data
    scaler = StandardScaler()
    df[df.select_dtypes(include=[np.number]).columns] = scaler.fit_transform(df.select_dtypes(include=[np.number]))
    
    # 6. PCA (if components are specified)
    if pca_components:
        pca = PCA(n_components=pca_components)
        pca_data = pca.fit_transform(df.select_dtypes(include=[np.number]))
        df_pca = pd.DataFrame(pca_data, columns=[f"PCA_{i+1}" for i in range(pca_components)])
        
        # Print explained variance
        print("Explained Variance Ratio:", pca.explained_variance_ratio_)

        # Compute feature contributions (PCA loadings)
        loadings = pd.DataFrame(pca.components_, columns=df.select_dtypes(include=[np.number]).columns, 
                                index=[f"PCA_{i+1}" for i in range(pca_components)])
        print("\nFeature Contributions to Each Principal Component:")
        print(loadings.T)  # Transpose for better readability

        df = df_pca  # Use PCA-transformed data

    # 7. Remove outliers using Z-score
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    z_scores = np.abs((df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std())
    df = df[(z_scores < z_threshold).all(axis=1)]
    
    return df
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

def preprocess_categorical(df, country_column="B_COUNTRY", fill_missing=True):
    df = df.copy()
    # Обрабатываем категориальные данные
    categorical_cols = df.select_dtypes(include=["object"]).columns
    for col in categorical_cols:
        if col == country_column:
            continue  # Пропускаем столбец со странами, чтобы сохранить их код

        if df[col].nunique() <= 10:  
            # Если <=10 уникальных значений, применяем Label Encoding
            df[col] = LabelEncoder().fit_transform(df[col])
        else:
            # Если больше, применяем частотное кодирование (frequency encoding)
            freq_map = df[col].value_counts(normalize=True).to_dict()
            df[col] = df[col].map(freq_map)

    # Заполняем пропущенные значения
    if fill_missing:
        for col in df.columns:
            if df[col].isnull().sum() > 0:
                if df[col].dtype == "object":
                    df[col].fillna(df[col].mode()[0], inplace=True)  # Заполняем модой
                else:
                    df[col].fillna(df[col].median(), inplace=True)  # Заполняем медианой


    # Создаем маппинг страны по коду
    country_mapping = {
        8: "Albania", 20: "Andorra", 32: "Argentina", 51: "Armenia", 36: "Australia",
        40: "Austria", 31: "Azerbaijan", 50: "Bangladesh", 112: "Belarus", 68: "Bolivia",
        70: "Bosnia Herzegovina", 76: "Brazil", 100: "Bulgaria", 124: "Canada", 152: "Chile",
        156: "China", 170: "Colombia", 191: "Croatia", 196: "Cyprus", 203: "Czechia",
        208: "Denmark", 218: "Ecuador", 818: "Egypt", 233: "Estonia", 231: "Ethiopia",
        246: "Finland", 250: "France", 268: "Georgia", 276: "Germany", 826: "Great Britain",
        300: "Greece", 320: "Guatemala", 344: "Hong Kong SAR", 348: "Hungary", 352: "Iceland",
        356: "India", 360: "Indonesia", 364: "Iran", 368: "Iraq", 380: "Italy",
        392: "Japan", 400: "Jordan", 398: "Kazakhstan", 404: "Kenya", 417: "Kyrgyzstan",
        428: "Latvia", 422: "Lebanon", 434: "Libya", 440: "Lithuania", 446: "Macao SAR",
        458: "Malaysia", 462: "Maldives", 484: "Mexico", 496: "Mongolia", 499: "Montenegro",
        504: "Morocco", 104: "Myanmar", 528: "Netherlands", 554: "New Zealand", 558: "Nicaragua",
        566: "Nigeria", 807: "North Macedonia", 909: "Northern Ireland", 578: "Norway",
        586: "Pakistan", 604: "Peru", 608: "Philippines", 616: "Poland", 620: "Portugal",
        630: "Puerto Rico", 642: "Romania", 643: "Russia", 688: "Serbia", 702: "Singapore",
        703: "Slovakia", 705: "Slovenia", 410: "South Korea", 724: "Spain", 752: "Sweden",
        756: "Switzerland", 158: "Taiwan ROC", 762: "Tajikistan", 764: "Thailand", 788: "Tunisia",
        792: "Turkey", 804: "Ukraine", 840: "United States", 858: "Uruguay", 860: "Uzbekistan",
        862: "Venezuela", 704: "Vietnam", 716: "Zimbabwe"
    }
    
    # Добавляем столбец с названием страны
    df["Country_Name"] = df[country_column].map(country_mapping)

    return df

# Использование:
# df_transformed = preprocess_categorical(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

def segment_and_plot(df, n_clusters=5):
    df = df.copy()

    # 1️⃣ Separate 'Country_Name' (for labeling)
    country_col = None
    if "Country_Name" in df.columns:
        country_col = df["Country_Name"]
        df = df.drop(columns=["Country_Name"])

    # 2️⃣ Apply K-Means Clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    df["Cluster"] = kmeans.fit_predict(df)

    # 3️⃣ Merge back 'Country_Name'
    if country_col is not None:
        df["Country_Name"] = country_col.reset_index(drop=True)

    return df, kmeans
```

```python
def plot_clusters(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df["PCA_1"], y=df["PCA_2"], hue=df["Cluster"], palette="tab10", s=50, alpha=0.7)
    
    # Annotate some country names for reference
    for i in range(0, len(df), len(df) // 10):  # Annotate 10 countries randomly
        plt.text(df["PCA_1"].iloc[i], df["PCA_2"].iloc[i], df["Country_Name"].iloc[i], fontsize=9, alpha=0.7)

    plt.title("Cluster Visualization with Country Labels")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend(title="Cluster")
    plt.show()
```

```python
# 2. Perform clustering
df_clusters, kmeans_model = segment_and_plot(df_cleaned, n_clusters=15)

# 3. Visualize the clusters
plot_clusters(df_clusters)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
def apply_dbscan(df, eps=0.5, min_samples=10, country_column="Country_Name"):
    df_numeric = df.drop(columns=[country_column], errors="ignore")  # Drop non-numeric column

    scaler = StandardScaler()
    df_scaled = scaler.fit_transform(df_numeric)

    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    cluster_labels = dbscan.fit_predict(df_scaled)

    df_result = df.copy()
    df_result["Cluster"] = cluster_labels

    return df_result, dbscan
```

```python
def plot_dbscan(df, country_column="Country_Name"):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df["PCA_1"], y=df["PCA_2"], hue=df["Cluster"], palette="tab10", alpha=0.6)

    # Add country labels for some points
    sampled = df.sample(50)  # Show only some labels to avoid clutter
    for i, row in sampled.iterrows():
        plt.text(row["PCA_1"], row["PCA_2"], str(row[country_column]), fontsize=8, color="black")

    plt.title("DBSCAN Clustering with Country Labels")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend(title="Cluster")
    plt.show()
```

```python
df_dbscan, dbscan_model = apply_dbscan(df_cleaned, eps=2, min_samples=10)
plot_dbscan(df_dbscan)
plot_clusters(df_clusters)
```
